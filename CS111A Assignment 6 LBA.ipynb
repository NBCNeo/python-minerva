(b) As we saw in (a), the radius of the circle is given by $f$. Therefore, the formula for the area $A$ would be $A(x) = \pi \cdot f(x)^2$

(c) First we need to define the radius of the cylinder. Let's say we will take the right endpoint given the points on the $x$ axis after dividing it in segments of size $\Delta x$ - this means we are taking the radius of the outer circles of the shells. Given this, the radius of cylinder $i$ will be $f(i*\Delta x)$. The volume is then the area of cylinder $i$, which is the formula from (b), times the height which is $\Delta x$

(i) $$V_i= A_i \cdot \Delta x = \pi \cdot f(i \cdot \Delta x)^2 \cdot \Delta x $$

The approximation of the volume is the sum of the volumes of the $n$ slices:

(ii) $$\sum_{i=1}^{n} V_i= \sum_{i=1}^{n} \pi \cdot f(i \cdot \Delta x)^2 \cdot \Delta x =  \pi \cdot \Delta x \sum_{i=1}^{n} f(i \cdot \Delta x)^2 $$

Now, if we are trying to approximate the volume on the interval $[a,b]$, then $\Delta x$ would be dividing the interval in $n$ slices, thus $\Delta x = \frac{b-a}{n}$. Substituting in the formula from (ii) we would get that the approximation, using right endpoints, of volume of the solid of revolution in the interval $[a,b]$ is

$$\sum_{i=1}^{n} V_i= \pi \cdot \frac{b-a}{n} \sum_{i=1}^{n}f(i \cdot \frac{b-a}{n})^2  $$

(d) The greater the number of slices we take, the better the approximation we are going to have. Which means that the expression that fives us the exact volume of the solid is

$$V_{total} = \pi \cdot \lim_{n\to\infty} \Delta x \sum_{i=1}^{n} f(i \cdot \Delta x)^2  $$

(e) Now, let's say that $h(x) = f(x)^2$, which means $h(x) = (4-x^2)^2 =  16-8x^2+ x^4 $ After substituting in the formula above, we get

$$V_{total} = \pi \cdot \lim_{n\to\infty} \Delta x \sum_{i=1}^{n} h(i \cdot \Delta x) $$

And given that $f$ was polynomial, then $h$ is also a polynomial, which means it's also continuous (and, in particular, is continuous on the interval $[a,b]$), we know that 

$$ \int_{a}^{b} f(x)^2 dx = \int_{a}^{b} h(x) dx = \lim_{n\to\infty} \sum_{i=1}^{n} h(x_i^{*}) \Delta x $$

and in this case $x_i^{*} = i \cdot \Delta x$, thus we have 

$$V_{total} = \pi \cdot \lim_{n\to\infty} \Delta x \sum_{i=1}^{n} h(i \cdot \Delta x) = \pi \cdot\int_{a}^{b} 16-8x^2+ x^4 dx = \pi \cdot \left[\frac{x^5}{5}-\frac{8x^3}{3} + 16x\right]_a^b$$

(f) To calculate the volume in Figure 2, we just need to use the last formula for the interval $[0,2]$:

$$\pi \cdot \int_{0}^{2} 16-8x^2+ x^4 dx = \pi \cdot \left[\frac{x^5}{5}-\frac{8x^3}{3} + 16x\right]_0^2$$ 

and using the Fundamental Theorem of Calculus, the above is equal to

$$\pi \cdot (\frac{2^5}{5}-\frac{8(2)^3}{3} + 16(2)) - 0\pi =\frac{256}{15}\pi$$

(g) Yes, in this case we would need get an expression for $x$ as a function of $y$. From $y=4-x^2$, we solve for $x$ and we get $x = \sqrt{4-y}$, which means that in this case our function would be of the form $h(y) = \sqrt{4-y}$, which is the inverse of our $f(x) = 4-x^2$. Using this $h$ and following an analogous procedure as before we would get that the volume is

$$ \pi \cdot \int_{0}^{4} (\sqrt{4-y})^2 dy = \pi \cdot \int_{0}^{4} (4y) dy = \pi \cdot \left[4y - \frac{y^2}{2}\right]_0^4 = 8 \pi$$

Note: The interval in this case is $[0,4]$ because that's the domain of Figure 1 on the $y$ axis.

This means that the solid has a smaller volume when revolved around the $y$ axis than when revolved around the $x$ axis.

on the interval $[a,b]$ which is the given domain of $x$ and when revolved around the $y$ axis is

(a) It would look like a cylinder with a hollow cylinder with an outer radius of 1 and height of $f(1)$, which means that the surface area of the side is the circumference given by $2 \pi r = 2 \pi \cdot 1$, times the height. Therefore, the surface is

$$S_{r=1} = 2 \pi \cdot f(1) = 6 \pi$$

(b) Which can be generalized as

$$ S(x) = 2 \pi x \cdot f(x)$$

(c) To approximate the volume of the slices we need to multiply the surface area times the width, which is $\Delta x$. Therefore we have that that the approximate volume of a slice is

i) $$V_i = 2 \pi c_i \cdot f(c_i) \Delta x = 2 \pi (\Delta x \cdot i) f(\Delta x \cdot i) \Delta x  $$

where $c_i = \Delta x \cdot i$ is taking the outer circle of the slice to approximate the volume of each of the slices.

Finally if we take $n$ slices, the approximation of the volume is given by the sum of the volume of each of the $n$ slices

ii) $$V_{approx} = \sum_{i=1}^{n} V_i = \sum_{i=1}^{n} 2 \pi (\Delta x \cdot i) f(\Delta x \cdot i) \Delta x = 2 \pi \sum_{i=1}^{n} (\Delta x \cdot i) \cdot f(\Delta x \cdot i) \Delta x $$

(d) As stated before, the approximation gets better as $n$ gets larger. Therefore, the exact volume is given by the following limit

$$ V_{exact} = \lim_{n\to\infty} 2 \pi \sum_{i=1}^{n} (\Delta x \cdot i) \cdot f(\Delta x \cdot i) \Delta x  = 2 \pi \cdot \lim_{n\to\infty} \sum_{i=1}^{n} (\Delta x \cdot i) \cdot f(\Delta x \cdot i) \Delta x $$

(e) Now, let's say $h(x) = x \cdot f(x) = 4x - x^3$ so that $h(\Delta x \cdot i) = (\Delta x \cdot i) \cdot f(\Delta x \cdot i)$. Therefore we have

$$ V_{exact} = 2 \pi \cdot \lim_{n\to\infty} \sum_{i=1}^{n}  h(\Delta x \cdot i) \Delta x $$

And because $h$ is polynomial then it is continuous (in particular, it is continuous on any interval $[a,b]$), then we have the following

$$ \int_{a}^{b} x \cdot f(x) dx = \int_{a}^{b} h(x) dx = \lim_{n\to\infty} \sum_{i=1}^{n} h(x_i^{*}) \Delta x $$

where $x_i^{*} = \Delta x \cdot i$ in this case. Therefore, we have

$$ V_{exact} = 2 \pi  \lim_{n\to\infty} \sum_{i=1}^{n}  h(\Delta x \cdot i) \Delta x = 2 \pi \int_{a}^{b} (4x - x^3) dx = 2 \pi \cdot \left[2x^2-\frac{x^4}{4}\right]_a^b $$

And if we calculate the previous on the interval $[0,2]$ then we would get

$$2 \pi \cdot \left[2x^2-\frac{x^4}{4}\right]_0^2 = 2 \pi (2(2)^2-\frac{2^4}{4}) - 0\pi = 8\pi$$

As we can see, it is the same result that we got in A.1(g)

(f) Yes, we would need to take the inverse of $f$ in that case, just as we did with the inverse of $f$ in A.1(g), and we would find out that we get the same result gotten before of $\frac{256}{15}\pi$

(g) As we saw in (d) and (e), but now using $g(x)$ instead of $f(x)$, we have that the volume if we revolve around the $y$ axis is 

$$ V_{exact} = 2 \pi \cdot \lim_{n\to\infty} \sum_{i=1}^{n} (\Delta x \cdot i) \cdot g(\Delta x \cdot i) \Delta x = 2 \pi \cdot \int_{a}^{b} x \cdot g(x) dx$$

on the interval $[a,b]$ and if we revolve around the $x$ axis the volume is given by
