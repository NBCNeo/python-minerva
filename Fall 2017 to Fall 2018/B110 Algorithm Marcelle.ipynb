```python
def Qtourist():
    a = raw_input('Are you a tourist? ')
    while a != 'yes' and a != 'YES' and a != 'Yes' and a != 'no' and a != 'No' and a != 'NO':
        print('Invalid response, please answer with yes or no\n')
        a = raw_input('Are you a tourist? ')
    a = a == 'yes' or a == 'Yes' or a == 'YES'
    return a
    
def Qfirst():
    a = raw_input('Is it your first time in Korea? ')
    while a != 'yes' and a != 'YES' and a != 'Yes' and a != 'no' and a != 'No' and a != 'NO':
        print('Invalid response, please answer with yes or no\n')
        a = raw_input('Is it your first time in Korea? ')
    a = a == 'yes' or a == 'Yes' or a == 'YES'
    return a

def Qalone():
    a = raw_input('Are you travelling alone? ')
    while a != 'yes' and a != 'YES' and a != 'Yes' and a != 'no' and a != 'No' and a != 'NO':
        print('Invalid response, please answer with yes or no\n')
        a = raw_input('Are you travelling alone? ')
    a = a == 'yes' or a == 'Yes' or a == 'YES'
    return a

def Qfood():
    a = raw_input('Do you want food with your drink? ')
    while a != 'yes' and a != 'YES' and a != 'Yes' and a != 'no' and a != 'No' and a != 'NO':
        print('Invalid response, please answer with yes or no\n')
        a = raw_input('Do you want food with your drink? ')
    a = a == 'yes' or a == 'Yes' or a == 'YES'
    return a

def Qanimals():
    a = raw_input('Does your group like animals? ')
    while a != 'yes' and a != 'YES' and a != 'Yes' and a != 'no' and a != 'No' and a != 'NO':
        print('Invalid response, please answer with yes or no\n')
        a = raw_input('Does your group like animals? ')
    a = a == 'yes' or a == 'Yes' or a == 'YES'
    return a

def Qwork():
    a = raw_input('Do you wish to work at the cafe? ')
    while a != 'yes' and a != 'YES' and a != 'Yes' and a != 'no' and a != 'No' and a != 'NO':
        print('Invalid response, please answer with yes or no\n')
        a = raw_input('Do you wish to work at the cafe? ')
    a = a == 'yes' or a == 'Yes' or a == 'YES'
    return a

def Qstudent():
    a = raw_input('Are you a student? ')
    while a != 'yes' and a != 'YES' and a != 'Yes' and a != 'no' and a != 'No' and a != 'NO':
        print('Invalid response, please answer with yes or no\n')
        a = raw_input('Are you a student? ')
    a = a == 'yes' or a == 'Yes' or a == 'YES'
    return a

def Qmeeting():
    a = raw_input('Do you wish to have a meeting at the cafe? ')
    while a != 'yes' and a != 'YES' and a != 'Yes' and a != 'no' and a != 'No' and a != 'NO':
        print('Invalid response, please answer with yes or no\n')
        a = raw_input('Do you wish to have a meeting at the cafe? ')
    a = a == 'yes' or a == 'Yes' or a == 'YES'
    return a

def Qage():
    a = raw_input('What is your age? ')
    integer = len(a) > 0
    for ch in a:
        if ch not in '0123456789':
            integer = False 
    while not integer:
        a = raw_input('What is your age? ')
    return int(a)

if Qtourist(): #tourist
    if Qfirst(): #first-timer
        if Qalone(): #alone
            if Qfood(): #want food
                print('\nYour cafe is Beansbins Coffee!')
            else: #no food
                print('\nYou should go to The Flower Cafe!')
        else: #not alone
            if Qanimals(): #likes animals
                print('\nBlind Alley Cafe is the perfect place for you!')
            else: #no animals
                print('\nYour coffee shop is Ddo-ong :)')
    else: #not first time
        print("\nI bet you haven't been to Cafe Gondry ;)")
        
            
else: #not a tourist
    if Qstudent(): #student
        if Qwork(): #work
            print('\nCheck out Maxim Plant!')
        else: #not work
            print("\nLet's go to the Chef's Dream Bakery!")
            
    else: #not a student 
        age = Qage()
        if age <= 25: #age < 25
            if Qalone(): #alone
                print('\nAnthracite Coffee Roasters is the place to go!')
            else: #not alone
                print('\nGo to 10000 Lab Namib!')
        elif 26 <= age and age <= 55: #age 26-55
            if Qmeeting(): #wish to have meeting
                print('\nWe work cafe is perfect for you!')
            else: #no meeting
                print('\nFritz Coffee Company it is!')
        else: #age > 56
            print('\nYou will love Shin Old Tea House!')
```
